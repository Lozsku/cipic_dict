```python
import numpy as np

def convert_dict_to_lists(data):
    if isinstance(data, dict):
        return {k: convert_dict_to_lists(v) for k, v in data.items()}
    elif isinstance(data, np.ndarray):
        return data.tolist()
    else:
        return data
```
